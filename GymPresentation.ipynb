```python
# Libs:
import matplotlib.pyplot as plt
import gymnasium as gym
import numpy as np
```

```python
# Instanciando
env = gym.make("CartPole-v1", render_mode="rgb_array")
state, info = env.reset(seed=42)
print(state)
```

```python
# Function to Visualization:
def render():
  '''
    Function to render the environment.
  '''
  state_image = env.render()
  plt.imshow(state_image)
  plt.show()
```

```python
# Env class methods
for met in dir(env):
  if not (met.startswith("__") and met.endswith("__")):
    print(met)
```

```python
# vizualization:
render()
```

```python
# Performing Actions:
action=1 # moving right // 0: moving left
state, reward, terminated,_,_ = env.step(action)

print("State: ", state)
print("Reward: ", reward)
print("Terminated: ", terminated)
```

```python
while not terminated:
  action = 1 # Move to the right
  state, reward, terminated, _, _ = env.step(action)
  render()
```
